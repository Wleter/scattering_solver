```python
import matplotlib.pyplot as plt
import numpy as np
import utility
```

Single Channel

```python
data = np.loadtxt('../data/single_chan/wave_function.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")

ax.plot(data[:, 0], data[:, 1] / np.max(np.abs(data[:, 1])))
ax.plot(data[:, 0], data[:, 2] / (0.25 * np.abs(np.min(data[:, 2]))))
ax.set_ylim(-1.1, 1.1);
```

```python
data = np.loadtxt('../data/single_chan/propagation_distance.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("scattering length [bohr]")

ax.plot(data[:, 0], data[:, 1])
ax.plot(data[:, 0], data[:, 2]);
```

```python
data = np.loadtxt('../data/single_chan/mass_scaling.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("mass scaling factor")
ax.set_ylabel("scattering length [bohr]")

utility.plot_scattering(ax, data[:, 0], data[:, 1], data[:, 2])
ax.set_ylim(-1000, 1000)
```

```python
data = np.loadtxt('../data/single_chan/bound_diffs.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("energy [cm$^{-1}$]")
ax.set_ylabel("bound difference [a.u.]")
ax.set_ylim(-10, 10)

ax.plot(data[:, 0], data[:, 1])
ax.plot(data[:, 0], data[:, 2])
```

Two channel

```python
data = np.loadtxt('../data/two_chan/wave_function.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")

normalization = np.max(np.abs(data[:, 1]))
ax.plot(data[:, 0], data[:, 1] / normalization)
ax.plot(data[:, 0], data[:, 2] / normalization)
ax.set_ylim(-1.1, 1.1);
```

```python
data = np.loadtxt('../data/two_chan/mass_scaling.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("mass scaling factor")
ax.set_ylabel("scattering length [bohr]")

utility.plot_scattering(ax, data[:, 0], data[:, 1], data[:, 2])
ax.set_ylim(-1000, 1000);
```

```python
ns = [0, 1, 3, -1, -2, -5]

fig, ax = utility.plot()
ax.set_xlabel("position [bohr]")
ax.set_ylabel("wave function")
ax.set_ylim(-1.1, 1.1)

for n in ns:
    data = np.loadtxt(f'../data/single_chan/bound_wave_{n}.dat', delimiter='\t', skiprows=1)

    normalization = np.max(np.abs(data[:, 1]))
    ax.plot(data[:, 0], data[:, 1] / normalization)
```

```python
data = np.loadtxt('../data/two_chan/bound_diffs.dat', delimiter='\t', skiprows=1)

fig, ax = utility.plot()
ax.set_xlabel("energy [cm$^{-1}$]")
ax.set_ylabel("bound difference [a.u.]")
ax.set_ylim(-20, 20)

a = np.median(data[:, 1])
scaling = data[:, 2].max() / (100 * a)


ax.plot(data[:, 0], scaling * data[:, 1])
ax.plot(data[:, 0], data[:, 2])
```
